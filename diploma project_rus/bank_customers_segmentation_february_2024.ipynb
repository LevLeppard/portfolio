### Пропуски

Подсчитаем количество пропусков по столбцам датасета конструкцией isna().sum().

```python
df.isna().sum()
```

26 пропусков в столбце 'age' и 2295 пропусков в столбце 'balance'. Необходимо определить тип пропусков и возможные причины их образования.

Учитывая немалое количество пропусков нам необходимо выяснить каким образом данные с пропуском отличаются от полноценных данных. Для этого добавим в датасет 2 дополнительных столбца ('nans_age', 'nans_balance') со следующей кодировкой:
- "0" пропуска нет,
- "1" пропуск.

Создадим списки с индексами датасета, в которых есть пропуски по вышеуказанной логике.

```python
nans_age = df.query('age.isna()').index
nans_balance = df.query('balance.isna()').index
```

Сделаем соответствующие словари для удобства кодировки.

```python
nans_age_dict = dict.fromkeys(nans_age, 1)
nans_balance_dict = dict.fromkeys(nans_balance, 1)
```

Создадим новый пустой столбец и заполним пропуски в нём соответствующими кодами.

```python
df['nans_age'] = None
df['nans_age'] = df['nans_age'].fillna(nans_age_dict).fillna(0)
```

```python
df['nans_balance'] = None
df['nans_balance'] = df['nans_balance'].fillna(nans_balance_dict).fillna(0)
```

Исследуем как данные с пропусками в указанных столбцах отличаются от данных без пропусков, для этого рассчитаем показатели центральной тенденции по всем пользовательским признакам и свёдем их в сводные таблицы результатов.

```python
values = ['age', 'est_salary', 'balance', 'score', 'equity', 'products']
columns = ['nans_age', 'nans_balance']
func = {'age':'mean',
        'est_salary':'median',
        'balance':'median',
        'score':['mean', 'count'],
        'equity':'median',
        'products':'median'}

for c in df.columns:
    if c in columns:
        nans_analysis = df.pivot_table(columns=df[c], values=values, aggfunc=func).round(1)
        nans_analysis.columns = ['Без пропусков', 'С пропусками']
        nans_analysis['Разница'] = (nans_analysis.diff(axis=1).drop(labels='Без пропусков', axis=1))
        nans_analysis['Относительная разница, %'] = round((nans_analysis['Разница']/
                                                          nans_analysis['Без пропусков'])*100, 1)
        display('Сравнение показателей центральной тенденции для признаков с пропусками и без по '+c, nans_analysis)
```

Данные с пропусками по своим усредненным показателям сильно отличаются от данных без пропусков по следующим признакам:
- 'balance' (только данные с пропусками в столбце 'age'),
- 'age' (только данные с пропусками в столбце 'balance'),
- 'equty',
- 'est_salary',
- 'products'.

Характер отклонений от данных без пропусков в данных с пропусками по обоим столбцам 'age' и 'balance' схожий, что может указывать на их общую "природу" возникновения.  

Сравнение категориальных признаков в столбцах 'city', 'gender', 'credit_card', 'last_activity', 'churn' в разрезе наличия и отсутствия пропусков визулизируем на круговых диаграммах.

```python
features = ['city', 'gender', 'credit_card', 'last_activity', 'churn']
nans = ['(без проп)', '(с проп)']
fig, ax = plt.subplots(nrows=len(features), ncols=len(columns)*len(nans), figsize=(17,20))

for c in range(len(columns)):
    for f in range(len(features)):
        for n in range(len(nans)):
            data=df.pivot_table(index=features[f], values='score', columns=columns[c], aggfunc='count')[n]
            ax[f,(c+n+c)].pie(data, labels=data.index, autopct='%1.1f%%')
            ax[f,(c+n+c)].set_title(columns[c]+': '+features[f]+' - '+nans[n])
plt.show()
```

В том числе и по категориальным признакам пропуски в данных в столбцах 'age' и 'balance' распределены похожим образом, некоторые отличия в распределении по признакам 'city', 'last_activity' и 'churn' скорее всего обусловлены тем, что пропусков в столбце 'age' всего 26, при этом в столбце 'balance' их гораздо больше и они должны лучше отражать отличия от данных без пропусков. Таким образом предположение о том, что причина появления пропусков в столбцах 'age' и 'balance' общая, имеет под собой основания.

Разброс значений в столбцах 'balance' и 'est_salary' очень большой, отчего много выбросов. Оценим их количество.

```python
def lower_upper(df):
    q1 = df.quantile(q=0.25)
    q3 = df.quantile(q=0.75)
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    return lower_bound, upper_bound
```

```python
total = []
for col in df.columns:
    if col in columns:
        lower_bound, upper_bound  = lower_upper(df[col])
        print(len(df[(df[col] < lower_bound) | (df[col] > upper_bound)]), ' строк со значениями-выбросами в столбце '+col)
        total.append(len(df[(df[col] < lower_bound) | (df[col] > upper_bound)]))
print('ИТОГО ЗНАЧЕНИЙ-ВЫБРОСОВ:', sum(total), 'шт. -', round(sum(total)/len(df)*100), '%')
```

14% датасета - выбросы (аномально высокие или аномально низкие значения), это максимально возможное значение, т.к. тут не учтена возможность, что выбросы по разным признакам будут "пересекаться".

```python
barplots = ['age', 'equity', 'products', 'credit_card', 'last_activity']
barplots_rus = ['возраст', 'объектов собственности', 'банковских продуктов', 'кредитная карта', 'активность']
genders = ['М', 'Ж', df['gender']]
genders_rus = ['МУЖЧИН', 'ЖЕНЩИН', '']
avg_churn = [avg_churn_male, avg_churn_female, avg_churn_total]
avg_churn_desc = [['средний % оттока у мужчин'], ['средний % оттока у женщин'], ['средний % оттока по банку']]
category = ['НЕТ', 'ДА']
colors = ['b', 'r', 'g']
colors_leg = ['r', 'g', 'b']

fig, ax = plt.subplots(nrows=len(barplots), ncols=len(genders), figsize=(15, 25))
for p in range(len(barplots)):
    for g in range(len(genders)):
        data = df.loc[df['gender'] == genders[g]].groupby(barplots[p])['churn'].mean()
        names = data.index
        values = data.values
        if max(names) == 1:
            ax[p,g].bar(data=data, x=category, height=values, width=.2, color=colors[g])
        else:
            ax[p,g].bar(data=data, x=names, height=values, color=colors[g])
        ax[p,g].set_title('Отток '+genders_rus[g]+' по - '+barplots[p]+' ('+barplots_rus[p]+')')
        ax[p,g].set_xlabel(barplots_rus[p])
        ax[p,g].set_ylabel('% оттока')
        ax[p,g].axhline(y=(avg_churn[g]/100), color=colors_leg[g], linewidth=3, linestyle='-.')
        ax[p,g].legend(avg_churn_desc[g])
        if g > 0:
            ax[p,g].sharey(ax[p,g-1])
            ax[p,g].set_ylabel('')
        fig.tight_layout()
        fig.show();
```

### Проверка статистических гипотез
    
Исходя из анализа пользовательских признаков клиентской базы сформулируем статистические гипотезы для последующей проверки:

- средний доход клиентов жещин ('est_salary'), попадающих в отток, больше аналогичного показателя остальных клиентов женщин;

**Гипотеза: средний отток клиентов банка одинаковый в Ярославле и Ростове Великом**

**тип:** двусторонняя

**метод:** z-test для пропорций, так как значения сравниваемых признаков бинарное.

На самом деле придется проверить как минимум 2 гипотезы, т.к. планируемая к использованию библиотека не имеет возможности использовать больше 2х выборок при сравнении. Сформулируем нулевую и альтернативную гипотезы:
- нулевая гипотеза: средние оттоки клиентов банка в Ярославле и Ростове Великом равны.
- альтернативная гипотеза: средние оттоки клиентов банка в Ярославле и Ростове Великом различаются.

Воспользуемся методом proportions_ztest() библиотеки statsmodels.

```python
# количество клиентов, попадающих в отток в городах
count = [df.query('city == "Ярославль" & churn == 1')['churn'].count(),
         df.query('city == "Ростов" & churn == 1')['churn'].count()]

# общее количество клиентов в городах
nobs = [df.query('city == "Ярославль"')['churn'].count(),
         df.query('city == "Ростов"')['churn'].count()]

# статистическую значимость установим в 5%
alpha = 0.05

stat, pval = proportions_ztest(count, nobs)

print('p-значение:', pval)

if pval < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Действительно, средний показатель оттока в Ярославле и Ростове Великом не отличается. Выведем также средние значения оттока для  клиентов из Ярославля и Ростова.

```python
print('средний отток клиентов в Ярославле:', df.query('city == "Ярославль"')['churn'].mean().round(3))
print('средний отток клиентов в Ростове:', df.query('city == "Ростов"')['churn'].mean().round(3))
```

**тип:** двусторонняя

**метод:** z-test для пропорций, так как значения сравниваемых признаков бинарное.

На самом деле придется проверить как минимум 2 гипотезы, т.к. планируемая к использованию библиотека не имеет возможности использовать больше 2х выборок при сравнении. Сформулируем нулевую и альтернативную гипотезы:
- нулевая гипотеза: средние оттоки клиентов банка в Ярославле и Рыбинске равны.
- альтернативная гипотеза: средние оттоки клиентов банка в Ярославле и Рыбинске различаются.

Воспользуемся методом proportions_ztest() библиотеки statsmodels.

```python
count = [df.query('city == "Ярославль" & churn == 1')['churn'].count(),
         df.query('city == "Рыбинск" & churn == 1')['churn'].count()]
nobs = [df.query('city == "Ярославль"')['churn'].count(),
         df.query('city == "Рыбинск"')['churn'].count()]

# статистическую значимость установим в 5%
alpha = 0.05

stat, pval = proportions_ztest(count, nobs)

print('p-значение:', pval)

if pval < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Отток в Рыбинске всё-таки отличается от Ярославля и, соответственно Ростова, отвергаем нулевую гипотезу в пользу альтернативной. Выведем также средние значения оттока для  клиентов из Ярославля и Рыбинска.

```python
print('средний отток клиентов в Ярославле:', df.query('city == "Ярославль"')['churn'].mean().round(3))
print('средний отток клиентов в Рыбинске:', df.query('city == "Рыбинск"')['churn'].mean().round(3))
```

В итоге, гипотезы о равенстве оттоков в разрезе городов не подтвердились, отток одинаковый только в Ярославле и Ростове, в Рыбинске судя по всему показатель меньше. Возможно распределение будущих сегментов клиентской базы в разрезе городов имеет под собой основание, проверим дополнительно этот момент, когда будем заниматься сегментацией.

**Гипотеза: средний остаток средств клиентов на счёте ('balance'), попадающих в отток, больше аналогичного показателя остальных клиентов.**

**Тип:** односторонняя вправо.

**Метод:** scipy.stats.ttest_ind(array1, array2, equal_var, alternative), так как у нас 2 независимые выборки из одной генеральной совокупности, при этом размер выборок разный, количество "отточных" клиентов значительно меньше, таким образом дисперсии у выборок могут быть разные (параметр equal_var=False).

Сформулируем нулевую и альтернативную гипотезы:
- нулевая гипотеза: средний остаток средств клиентов (мужчин и женщин) на счёте ('balance'), попадающих в отток, равно аналогичному показателю у остальных клиентов.
- альтернативная гипотеза: средний остаток средств клиентов (мужчин и женщин) на счёте ('balance'), попадающих в отток, больше аналогичного показателя остальных клиентов.

При проверке гипотез с использованием ttest необходимо избавиться от значений-выбросов, чтобы они не искажали среднеарифметическое значение, а как мы помним столбец 'balance' имеет таковые.

```python
balance_1 = df.query('churn == 1 & balance.notna() & balance <= @upper_bound_1')['balance']
balance_2 = df.query('churn == 0 & balance.notna() & balance <= @upper_bound_1')['balance']

# статистическую значимость установим в 5%
alpha = 0.05

results = st.ttest_ind(balance_1, balance_2, equal_var=False, alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Действительно, ttest подтверждает нашу гипотезу: средний остаток средств клиентов на счёте ('balance'), попадающих в отток, больше аналогичного показателя остальных клиентов. Данный признак, судя по всему, подходит нам для дальнейшей сегментации клиентской базы. Выведем также средние значения признака 'balance' для "отточных" и "неотточных" клиентов.

## Сегментация клиентов

Учитывая немалое количество клиентских признаков, по которым можно провести сегментацию, мы обратимся к инструментарию машинного обучения для кластеризации клиентской базы, с целью последующего использования сформированных кластеров для формирования клиентских сегментов.

В качестве инструмента машинного обучения мы используем агломеративную иерархическую кластеризацию из библиотеки scipy как способ получения оптимального количества кластеров. Затем мы применим алгоритм KMeans() из библиотеки sklearn как способ проверки.

Сначала создадим отдельный датасет для использования при сегментации и удалим из него клиентские признаки, которые определились не использовать.

```python
df_segment = df.drop(labels=['user_id', 
                             'city',
                             'balance',
                             'balance_salary',
                             'credit_card', 
                             'nans_balance',
                             'gender',
                             'churn'], axis=1)
```

Стандартизуем данные перед работой с алгоритмами.

```python
sc = StandardScaler()
x_sc = sc.fit_transform(df_segment)
```

Создадим объект linkage() для отрисовки иерархической кластеризации на дендрограмме.

```python
linked = linkage(x_sc, method = 'ward')
```

Построим дендрограмму и укажем, по нашему мнению, оптимальное количество кластеров, проведя красную пунктирную горизонтальную линию.

```python
plt.figure(figsize=(14, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация клиентов банка')
plt.axhline(y=125, color='red', linestyle='--')
plt.legend(['оптимальное количество кластеров'], title='', loc="best")
plt.xticks(ticks=[])
plt.show();
```

Красная линия добавлена нами вручную и пересекает основания 4 кластеров. В данном случае кластеры имеют чёткую границу друг с другом и дендрограмма с методом linkage() предлагает нам также 4 (разделены по цвету).

Теперь используем метрику silhouette_score() на различных "прогонах" алгоритма kMeans() задавая разное количество кластеров вручную, затем построим график полученных значений silhouette_score(), чтобы понять какое количество кластеров лучше всего подчёркивает различия между ними. Значение метрики silhouette_score() принимает значения от -1 до 1. Чем ближе к 1, тем качественнее кластеризация. 

```python
clusters = range(2, 12)

silhouette_scores = []
cluster = []

for c in clusters:
    km = KMeans(n_clusters = c, random_state=0)
    labels = km.fit_predict(x_sc)
    silhouette_scores.append(silhouette_score(x_sc, labels))
    cluster.append(c)
```

```python
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(cluster, silhouette_scores, 'o', color='tab:brown')
ax.plot(cluster, silhouette_scores)
plt.axhline(y=max(silhouette_scores), color='red', linestyle='--')
plt.title('Значения silhouette_score на различном количестве кластеров по методу kMeans()')
plt.xlabel('количество кластеров')
plt.ylabel('silhouette_score');
```

Агломеративная кластеризация и метод kMeans() показывают нам немного различные результаты (агломерация - 4, kMeans() - 6). Значение silhouette_score() для 6 максимальный.

Посмотрим на меры центральной тенденции по признакам в разрезе сформированных алгоритмом kMeans() 6 кластеров.

Запустим еще раз  алгоритм KMeans() для 6 кластеров (с наибольшим значением silhouette_score()), чтобы сформировать метки кластеров для датасета и создадим в нашем датасете столбец с признаком кластера, построим сводную таблицу с группировкой признаков по кластерам, отсортированную по величине оттока в порядке убывания.
