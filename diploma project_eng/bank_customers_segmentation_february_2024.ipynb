```python
values = ['age', 'est_salary', 'balance', 'score', 'equity', 'products']
columns = ['nans_age', 'nans_balance']
func = {'age':'mean',
        'est_salary':'median',
        'balance':'median',
        'score':['mean', 'count'],
        'equity':'median',
        'products':'median'}

for c in df.columns:
    if c in columns:
        nans_analysis = df.pivot_table(columns=df[c], values=values, aggfunc=func).round(1)
        nans_analysis.columns = ['without omissions', 'with omissions']
        nans_analysis['Difference'] = (nans_analysis.diff(axis=1).drop(labels='without omissions', axis=1))
        nans_analysis['Relative differrence, %'] = round((nans_analysis['Difference']/
                                                          nans_analysis['without omissions'])*100, 1)
        display('Comparison of measures of central tendency for traits with and without omissions by '+c, nans_analysis)
```

The variation of values in the 'balance' and 'est_salary' columns is very large, which results in a lot of outliers. Let's estimate their number.

```python
def lower_upper(df):
    q1 = df.quantile(q=0.25)
    q3 = df.quantile(q=0.75)
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    return lower_bound, upper_bound
```

```python
total = []
for col in df.columns:
    if col in columns:
        lower_bound, upper_bound  = lower_upper(df[col])
        print(len(df[(df[col] < lower_bound) | (df[col] > upper_bound)]), ' rows with outliers in the column '+'"'+col+'"')
        total.append(len(df[(df[col] < lower_bound) | (df[col] > upper_bound)]))
print('total number of outliers:', sum(total), ' -', round(sum(total)/len(df)*100), '% of the dataset')
```

14% of the dataset - outliers (abnormally high or abnormally low values), this is the maximum possible value, because it does not take into account the possibility that outliers by different characteristics will "overlap".

Abnormal values in the 'balance' and 'est_salary' columns are enough for a separate segment (about 9% of the dataset).

## Hypothesis testing and conclusions from EDA results

#### Statistical Hypothesis Testing
    
Based on the analysis of user attributes of the client base, let's formulate statistical hypotheses for further testing:
    
- the average outflow of bank customers is the same in Yaroslavl and Rostov Veliky,
- average outflow of bank customers is the same in Yaroslavl and Rybinsk,
- the average balance of customers (male and female) in the account ('balance') falling into the churn rate is higher than that of other customers;
- the average number of customers' real estate properties ('equity') falling into outflow is higher than that of other customers;
- the average balance/salary ratio of outflow customers ('balance/salary') is higher than that of other customers; 
- the average income of female customers ('est_salary') falling into churn is higher than that of other female customers;
- the average income of male customers ('est_salary') falling into churn is equal to that of other male customers;
- the average value of 'credit_card' attribute for women falling into churn is equal to 0.4;
- the average value of credit scoring ('score') of customers falling into churn is greater than that of other customers;
- the average value of 'last_activity' attribute for churned customers is 0.7;
- the average number of banking products ('products') of churn customers is higher than that of other customers.

**Hypothesis: average bank customer churn is the same in Yaroslavl and Rostov Veliky**

**type:** bilateral

**method:** z-test for proportions, since the values of the compared features are binary.

In fact, we will have to test at least 2 hypotheses, because the library planned for use does not have the possibility to use more than 2 samples in the comparison. Let's formulate the null and alternative hypotheses:
- null hypothesis: average outflows of bank customers in Yaroslavl and Rostov Veliky are equal.
- alternative hypothesis: average outflows of bank customers in Yaroslavl and Rostov Veliky differ.

Let's use the proportions_ztest() method of the statsmodels library.

```python
# number of customers falling into churn in cities
count = [df.query('city == "Yaroslavl" & churn == 1')['churn'].count(),
         df.query('city == "Rostov" & churn == 1')['churn'].count()]

# total number of customers in cities
nobs = [df.query('city == "Yaroslavl"')['churn'].count(),
         df.query('city == "Rostov"')['churn'].count()]

# statistical significance will be set at 5%
alpha = 0.05

stat, pval = proportions_ztest(count, nobs)

print('p-value:', pval)

if pval < alpha:
    print('Rejecting the null hypothesis')
else:
    print('Failed to reject the null hypothesis') 
```

Indeed, the average churn rate in Yaroslavl and Rostov Veliky does not differ. Let us also derive average churn rates for customers from Yaroslavl and Rostov.

```python
print('average churn in Yaroslavl:', df.query('city == "Yaroslavl"')['churn'].mean().round(3))
print('average churn in Rostov:', df.query('city == "Rostov"')['churn'].mean().round(3))
```

**type:** bilateral

In fact, we will have to test at least 2 hypotheses, since the library planned for use does not have the possibility to use more than 2 samples in the comparison. Let's formulate the null and alternative hypotheses:

## Customer segmentation

Given the considerable number of customer attributes on which segmentation can be performed, we turn to machine learning tools to cluster the customer base, in order to then use the generated clusters to form customer segments.

We will use agglomerative hierarchical clustering from the scipy library as a machine learning tool as a way to obtain the optimal number of clusters. Then we will apply the KMeans() algorithm from the sklearn library as a way of validation.

First, we will create a separate dataset for use in segmentation and remove client attributes from it that are determined not to be used.

```python
df_segment = df.drop(labels=['user_id', 
                             'city',
                             'balance',
                             'balance_salary',
                             'credit_card', 
                             'nans_balance',
                             'gender',
                             'churn'], axis=1)
```

We standardize data before working with algorithms.

```python
sc = StandardScaler()
x_sc = sc.fit_transform(df_segment)
```

Let's create a linkage() object to render hierarchical clustering on the dendrogram.

```python
linked = linkage(x_sc, method = 'ward')
```

We plot the dendrogram and indicate what we think is the optimal number of clusters by drawing a red dashed horizontal line.

```python
plt.figure(figsize=(14, 10))  
dendrogram(linked, orientation='top')
plt.title('Hierarchical clustering of bank customers')
plt.axhline(y=125, color='red', linestyle='--')
plt.legend(['optimal number of clusters'], title='', loc="best")
plt.xticks(ticks=[])
plt.show();
```

The red line is added manually and crosses the bases of 4 clusters. In this case, the clusters have a clear boundary with each other and the dendrogram with the linkage() method offers us also 4 (separated by color).

Now we use the silhouette_score() metric on different "runs" of the kMeans() algorithm by setting different numbers of clusters manually, then plot the resulting silhouette_score() values to see which number of clusters best emphasizes the differences between them. The value of the silhouette_score() metric takes values from -1 to 1. The closer to 1, the better the clustering.

```python
clusters = range(2, 12)

silhouette_scores = []
cluster = []

for c in clusters:
    km = KMeans(n_clusters = c, random_state=0)
    labels = km.fit_predict(x_sc)
    silhouette_scores.append(silhouette_score(x_sc, labels))
    cluster.append(c)
```

```python
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(cluster, silhouette_scores, 'o', color='tab:brown')
ax.plot(cluster, silhouette_scores)
plt.axhline(y=max(silhouette_scores), color='red', linestyle='--')
plt.title('Values of silhouette_score on different number of clusters by kMeans() method')
plt.xlabel('number of clusters')
plt.ylabel('silhouette_score');
```

Agglomerative clustering and the kMeans() method show us slightly different results (agglomerative is 4, kMeans() is 6). The silhouette_score() value for 6 is maximized.

Let's look at the measures of central tendency by trait in terms of the 6 clusters generated by the kMeans() algorithm.

Let's run the kMeans() algorithm again for 6 clusters (with the highest silhouette_score() value) to generate cluster labels for the dataset and create a column with the cluster feature in our dataset, build a summary table with clustering of features by clusters, sorted by outflow value in descending order.

```python
km = KMeans(n_clusters = 6, random_state=0)
labels = km.fit_predict(x_sc)
df_clusters = df
df_clusters['cluster'] = labels
```

```python
values = ['user_id', 'age', 'gender', 'city', 'est_salary', 'balance', 'score', 'equity', 'products',
          'credit_card', 'last_activity', 'balance_salary', 'churn']
func = {'churn':'mean',
        'age':['min', 'mean', 'max'],
        'est_salary':['min', 'median', 'max'],
        'balance':['min', 'median', 'max'],
        'score':['min', 'mean', 'max'],
        'equity':['min', 'median', 'max'],
        'products':['min', 'median', 'max'],
        'balance_salary':['min', 'median', 'max'],
        'user_id':'count',
        'credit_card': st.mode,
        'last_activity': st.mode,
        'gender': st.mode,
        'city': st.mode}
pt_clusters = (
    df_clusters.pivot_table(index='cluster', values=values, aggfunc=func).
    round(2).T.
    sort_values(by=[('churn', 'mean')], axis=1, ascending=False)
)
pt_clusters
```
